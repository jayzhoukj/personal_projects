```python
import math, os, numpy as np, pandas as pd, matplotlib.pyplot as plt
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
```

```python
df = pd.read_excel("FSL_experimental.xlsx", 
                   header=0, 
                   sheet_name=0)
df.head()
```

```python
def check_na(x):
    if type(x) is str:
        return True
    else:
        return not(math.isnan(x))

df['Bool'] = df['Category'].apply(lambda x: check_na(x))
df = df.loc[df.Bool,'Artist':'Year'].reset_index(drop=True)
df.head()
```

```python
df_spotify = pd.read_csv("SpotifyAudioFeaturesApril2019.csv",
                         header=0)
df_spotify.head(20)
```

```python
df2 = pd.merge(df, df_spotify, left_on=['Artist','Songs'], right_on=['artist_name', 'track_name'], how='left')
df2.tail()
```
