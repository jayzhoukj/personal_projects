```python
pag.click(10, 1070)
pag.typewrite([char for char in 'bookshelf'] + ['enter'], interval=0.1)

time.sleep(3)
pag.click(200, 200)

pages = ['C', 'C', 
         'i', 'ii', 
         'iii', 'iv', 
         'v', 'vi', 
         'vii', 'viii', 
         'ix', 'x', 
         'xi', 'xii', 
         'xiii', 'xiv', 
         'xv', 'xv'] + [str(i) for i in range(1,199)]

time.sleep(5)

pag.click(1682, 990)

time.sleep(2)

for i in range(int(len(pages)/2)):
    lower_limit = pages[2*i]
    upper_limit = pages[2*i + 1]
    
    pag.click(795, 462)
    pag.hotkey('ctrl', 'a')
    pag.typewrite(lower_limit)
    
    pag.moveRel(200, 0, duration=0.2)
    pag.click()
    pag.hotkey('ctrl', 'a')
    pag.typewrite(upper_limit)
    
    pag.moveRel(125, 0, duration=0.5)
    pag.click()
    
    time.sleep(28)
    
    pag.click(143, 965)
    
    time.sleep(3)
    
    pag.moveTo(256, 854, duration=0.2)
    pag.click()
    pag.typewrite(str(i).zfill(2))
    
    pag.moveTo(1669, 974, duration=0.2)
    pag.click()
    
    time.sleep(5)
```

```python
# A generator function 
def simpleGeneratorFun(): 
    yield 1
    yield 2
    yield 3
   
# x is a generator object
x = simpleGeneratorFun() 
  
# Iterating over the generator object using next 
print(x.__next__()) # In Python 3, __next__() 
print(x.__next__()) 
print(x.__next__())
```
